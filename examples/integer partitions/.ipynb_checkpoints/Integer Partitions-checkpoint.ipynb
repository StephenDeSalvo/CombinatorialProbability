```python
from scipy.stats import geom, uniform
```

```python
partition = {}
counts = 0
keep_going = True
n = ip.target['n']
x = ip.x_
while keep_going is True:
    partition = {i:y for i, y in enumerate([geom.rvs(1-x**i)-1 for i in range(2, n+1) if x**i != 1.0]) if y != 0}
    U = uniform()
    if U < geom.pmf(n - numpy.sum([x*y for x,y in partition.items()]), 1-x) / geom.pmf(0, 1-x):
        keep_going = False
    counts += 1
```

```python
partition = {}
counts = 0
keep_going = True
while keep_going is True:
    partition = {(i+2):y for i, y in enumerate([geom.rvs(1-x**i)-1 for i in range(2, n+1) if x**i != 1.0]) if y != 0}
    U = uniform().rvs(size=1)
    residual = n - numpy.sum([x*y for x,y in partition.items()])
    if U < geom.pmf(residual, 1-x, -1) / geom.pmf(0, 1-x, -1):
        keep_going = False
        if residual > 0:
            partition[1] = residual
    counts += 1
```
